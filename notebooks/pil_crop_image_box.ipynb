```python
# line coordinates
line_num = 8

df_line = df[df['line_num'] == line_num]
df_line = df_line.dropna()

left_a = df_line['left'].to_list()
top_a = df_line['top'].to_list()
width_a = df_line['width'].to_list()
height_a = df_line['height'].to_list()

x0 = left_a[0]
y0 = top_a[0]
x1 = left_a[-1] + width_a[-1]
y1 = top_a[-1] + height_a[-1]

print(x0, y0, x1, y1)
cropped_img = img.crop((x0, y0, x1, y1))
cropped_img.show()
cropped_img.save(f'../data/pdf_processed/101-kermokue-kh-adyge-khybaryzhkher.pdf/line_box_raw/{img_name}__{line_num}.jpg')
```

```python
with open(f'../data/pdf_processed/101-kermokue-kh-adyge-khybaryzhkher.pdf/line_box_raw/{img_name}__{line_num}.txt', 'w', encoding='utf-8') as f:
    f.write(' '.join(df_line['text'].to_list()))
```

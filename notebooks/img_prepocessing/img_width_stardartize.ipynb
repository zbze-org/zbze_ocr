```python
import os.path

from PIL import Image

files_dir = '../../data/pdf_processed/100-mafiedz-serebii-tkhygekher-iape-tom-povestkher-1995ge.pdf_unpapered'
imgs = os.listdir(files_dir)
```

```python
STANDARD_WIDTH = 2481

new_dir = os.path.join(files_dir, 'standardized')
os.makedirs(new_dir, exist_ok=True)


def standardize_img_width(file_path, output_dir=None, standard_width=STANDARD_WIDTH):
    if output_dir is None:
        output_dir = os.path.dirname(file_path)
    os.makedirs(output_dir, exist_ok=True)

    image = Image.open(file_path)
    width, height = image.size

    if width < standard_width:
        # add white background to the left and right of the image
        new_image = Image.new('RGB', (standard_width, height), color='white')
        offset = (standard_width - width) // 2
        new_image.paste(image, (offset, 0))
    elif width > standard_width:
        # crop the image
        offset = (width - standard_width) // 2
        new_image = image.crop((offset, 0, offset + standard_width, height))
    else:
        new_image = image

    output_path = os.path.join(output_dir, os.path.basename(file_path))
    new_image.save(output_path)


for img in imgs:
    img_path = os.path.join(files_dir, img)
    standardize_img_width(img_path, new_dir)
```
